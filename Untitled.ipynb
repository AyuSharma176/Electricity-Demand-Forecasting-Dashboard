```python
# Step 4: Load & Inspect Dataset
import pandas as pd

# 👇 update with your actual filename in Downloads/data
FILE_PATH = "/home/monarch/Downloads/data/.csv"

df = pd.read_csv(FILE_PATH)

# show first 5 rows
print("First 5 rows:")
display(df.head())

# show column names
print("\nColumn names:")
print(df.columns.tolist())

# check basic info
print("\nDataset Info:")
print(df.info())

# quick stats
print("\nSummary Stats:")
display(df.describe())
```

```python
# Step 5: Preprocessing & Feature Engineering

# convert Datetime column to actual datetime
df['Datetime'] = pd.to_datetime(df['Datetime'])

# set as index (important for time series)
df = df.set_index('Datetime')

# check for missing values
print("Missing values per column:")
print(df.isnull().sum())

# create calendar/time-based features
df['hour'] = df.index.hour
df['day'] = df.index.day
df['dayofweek'] = df.index.dayofweek  # 0 = Monday, 6 = Sunday
df['month'] = df.index.month
df['year'] = df.index.year
df['is_weekend'] = (df['dayofweek'] >= 5).astype(int)  # Sat/Sun = weekend

# show updated dataframe
print("\nUpdated dataframe with features:")
display(df.head())
```

```python
# Step 6: Train-Test Split
import numpy as np

# 80% train, 20% test (time-based split)
train_size = int(len(df) * 0.8)
train, test = df.iloc[:train_size], df.iloc[train_size:]

print("Train shape:", train.shape)
print("Test shape:", test.shape)

# features (X) and target (y)
FEATURES = ['hour', 'day', 'dayofweek', 'month', 'year', 'is_weekend']
TARGET = 'AEP_MW'

X_train, y_train = train[FEATURES], train[TARGET]
X_test, y_test = test[FEATURES], test[TARGET]
```

```python
# Step 7: Scale features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Step 8 (lighter version): Quick Training
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

# smaller model for faster run
model = XGBRegressor(
    n_estimators=100,   # fewer trees for speed
    learning_rate=0.1,
    max_depth=4,
    random_state=42
)

print("⏳ Training...")
model.fit(X_train_scaled, y_train)

print("✅ Training Done! Predicting...")
y_pred = model.predict(X_test_scaled)

# evaluate
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)

print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")
```

```python
# demand_forecast.py

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from xgboost import XGBRegressor

# 1. Load dataset
df = pd.read_csv("data/AEP_hourly.csv", parse_dates=["Datetime"])
df = df.set_index("Datetime")

# 2. Feature engineering - create lag features
for lag in [1, 24, 168]:  # 1 hour ago, 1 day ago, 1 week ago
    df[f"lag_{lag}"] = df["AEP_MW"].shift(lag)

# Drop missing rows (from lag creation)
df = df.dropna()

# 3. Define features and target
X = df.drop("AEP_MW", axis=1)
y = df["AEP_MW"]

# 4. Train-test split (80-20)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False
)

# 5. Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 6. Train model
print("⏳ Training...")
model = XGBRegressor(n_estimators=500, learning_rate=0.05, random_state=42)
model.fit(X_train_scaled, y_train)

# 7. Predictions
print("✅ Training complete! Predicting...")
y_pred = model.predict(X_test_scaled)

# 8. Evaluation
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)
print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")

# 9. Visualization - actual vs predicted
plt.figure(figsize=(12, 6))
plt.plot(y_test.index[:200], y_test[:200], label="Actual", color="blue")
plt.plot(y_test.index[:200], y_pred[:200], label="Predicted", color="red")
plt.title("Power Demand Prediction (First 200 test samples)")
plt.xlabel("Time")
plt.ylabel("MW")
plt.legend()
plt.show()
```
